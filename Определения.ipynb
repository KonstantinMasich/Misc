# Основные понятия
## Standard deviation - Среднеквадратичное отклонение
* **Обозначение:** $\mathbf{\sigma}$ (в популяции), **s** (в выборке)
* **Определение:** показатель рассеивания значений случайной величины относительно её математического ожидания
* **Суть:** показывает, насколько данные в наборе "скучкованы" возле среднего значения. Чем меньше $\sigma$, тем плотнее к среднему лежат данные, и *тем надёжнее само среднее*. Например, есть два менеджера, управляющие складом. Ниже представлены результаты их работы:
![](img2/f2.png)
Среднее у них примерно одинаково, то есть вроде бы они оба работают одинаково, но стандартное отклонение у первого менеджера составляет 2 (т.е. данные в среднем отклоняются от среднего значения на 2 единицы), что говорит о надёжности рассчитанного среднего и низкой вариативности данных. Это указывает на последовательность и надёжность работы первого менеджера: он выдаёт стабильные хорошие результаты. У второго стандартное отклонение составляет 18,9 (т.е. данные в среднем отклоняются от среднего значения на 18.9 единиц), что говорит о ненадёжности рассчитанного среднего: по мере добавления новых данных, среднее у второго менеджера будет сильно "прыгать" вверх-вниз, тогда как у первого оно будет гораздо более стабильно, а следовательно, результат его работы более предсказуем и в целом он работает лучше.
* **Расчёт:** $\sigma=\sqrt{\frac{\sum_{i=1}^{n}(x_i-\overline{x})^2}{n}}$ для всей популяции, 
$s=\sqrt{\frac{\sum_{i=1}^{n}(x_i-\overline{x})^2}{n-1}}$ для выборки, где $\mathsf{x}=(x_1, x_2,\dots, x_n)$

## Variance - Дисперсия
* **Обозначение:** **Var(x)**, $\mathbf{\sigma}^2$ (в популяции), $\mathbf{s^2}$ (в выборке)
* **Определение:** мера разброса значений случайной величины относительно её математического ожидания
* **Суть:** показывает, насколько данные в наборе "скучкованы" возле среднего значения. Является квадратом стандартной девиации и измеряется в квадратах единиц измерения, которые были изначально (если изначально были метры, то дисперсия измеряется в квадратных метрах).
* **Расчёт:** $Var(x)=\frac{\sum_{i=1}^{n}(x_i-\overline{x})^2}{n}$ для всей популяции, 
$s^2=\frac{\sum_{i=1}^{n}(x_i-\overline{x})^2}{n-1}$ для выборки, где $\mathsf{x}=(x_1, x_2,\dots, x_n)$

## Covariance - Ковариация
* **Обозначение:** $\mathbf{cov(X, Y)}$
* **Определение:** мера взаимной вариабельности двух случайных величин. Если бОльшие значения одной переменной соответствуют бОльшим значениям другой переменной, и то же самое касается меньших значений этих двух переменных, то их ковариация позитивна. В обратном случае, когда бОльшие значения одной переменной соответствуют меньшим значениям другой переменной, ковариация этих переменных негативна. Таким образом, знак ковариации показывает тенденцию в линейном отношении между переменными.
<br>Величину ковариации не так просто интерпретировать, поскольку она не нормализована и, соответственно, зависит от магнитуд переменных. Нормализованная версия ковариации, так называемый **коэффициент коррелляции**, как раз своей магнитудой показывает степень линейного соотношения между переменными (т.е. чем этот коэффициент больше, тем большая линейная зависимость существует между переменными).
